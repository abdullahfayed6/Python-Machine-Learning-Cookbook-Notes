# **Improving Performance Through Boosting**

You need a model with better performance than decision trees or random forests.

```python
from sklearn.ensemble import AdaBoostClassifier , AdaBoostRegressor
from sklearn import datasets
```

```python
# Load data
iris = datasets.load_iris()
features = iris.data
target = iris.target
```

---
AdaBoost is a powerful boosting technique that iteratively trains a series of weak models (often shallow decision trees, also called "stumps"). Each iteration emphasizes observations that were misclassified by the previous model.

---


```python
# Create adaboost tree classifier object
adaboost = AdaBoostClassifier(random_state=0)
# Train model
model = adaboost.fit(features, target)
```

Key Parameters

| Parameter             | Description |
|-----------------------|-------------|
| **`base_estimator`**  | The learning algorithm used to train the weak models. Usually a decision tree (`DecisionTreeClassifier` or `DecisionTreeRegressor`) by default. |
| **`n_estimators`**    | The number of weak models (iterations) to train. |
| **`learning_rate`**   | Controls how much each model contributes to the final prediction. Lower values slow learning but may improve generalization. |
| **`loss`** *(Regressor only)* | Specifies the loss function used when updating weights. Options: `"linear"` (default), `"square"`, or `"exponential"`. |
